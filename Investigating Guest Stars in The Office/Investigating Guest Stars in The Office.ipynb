<p>In this notebook, we will take a look at a dataset of The Office episodes, and try to understand how the popularity and quality of the series varied over time. To do so, we will use the following dataset: <code>datasets/office_episodes.csv</code>, which was downloaded from Kaggle <a href="https://www.kaggle.com/nehaprabhavalkar/the-office-dataset">here</a>.</p>
<p>This dataset contains information on a variety of characteristics of each episode. In detail, these are:

    <li><b>has_guests:</b> True/False column for whether the episode contained guest stars.</li>

```python
import matplotlib.pyplot as plt
import pandas as pd

office = pd.read_csv('datasets/office_episodes.csv')

print(office.info())
print(office.head())
```

```python
fig, ax = plt.subplots()

colors = []
for index, row in office.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append('red')
    elif row['scaled_ratings'] < 0.5 and row['scaled_ratings'] >= 0.25:
        colors.append('orange')
    elif row['scaled_ratings'] < 0.75 and row['scaled_ratings'] >= 0.5:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')

markersize = [250 if x==1 else 25 for x in office['has_guests'].values]
marker_style = ['*' if x==1 else '.' for x in office['has_guests'].values]

for num, view, color, size, style in zip(office['episode_number'],office['viewership_mil'],colors,markersize,marker_style):
    ax.scatter(num, view, c=color, s=size, marker=style)


ax.set_title('Popularity, Quality, and Guest Appearances on the Office')
ax.set_xlabel('Episode Number')
ax.set_ylabel('Viewership (Millions)')

plt.show()
```
