```python
from langdetect import detect
```

```python
other_languages_id = [3538, 1732, 2340, 4580, 2098, 4159, 76, 1407, 3972]
```

#### 2.1.1.1 Flan-T5 Model
model info: https://huggingface.co/google/flan-t5-small

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

T5Tokenizer requires the SentencePiece library

```python
pip install sentencepiece
```

```python
import sentencepiece
```

```python
model_name='google/flan-t5-small'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
for id in other_languages_id:
    transcript = transcripts[transcripts['id']==id]['transcript'].values[0]
    language = detect(transcript)
    input_text = f"translate {language} to en:\n {transcript}"
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    outputs = tokenizer.decode(
        model.generate(input_ids)[0],
        skip_special_tokens=True
        )

    # print prompt and translation
    print(transcript)
    print(f'Detected language: {language}')
    print('Translation:')
    print(outputs)
    print(dash_line)
```

* Error message: "Token indices sequence length is longer than the specified maximum sequence length for this model (612 > 512)."
* `google/flan-t5-small` has a maximum input sequence length of 512 tokens
* Conclusion: this model can't handle lengthy text unless we truncate a single transcript into several pieces to meet the sequence length limitations - I don't think it's neccessary as we can use a different model

#### 2.1.1.2 NLLB Model
model info: https://huggingface.co/facebook/nllb-moe-54b

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
# I used the distilled version: https://huggingface.co/facebook/nllb-200-distilled-600M
model_name='facebook/nllb-200-distilled-600M'

model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python
for id in other_languages_id:
    transcript = transcripts[transcripts['id']==id]['transcript'].values[0]
    source_lang = detect(transcript)
    input_text = f"translate {source_lang} to en:\n{transcript}"
    inputs = tokenizer(input_text, return_tensors="pt", padding=True)
    translated_tokens = model.generate(**inputs, forced_bos_token_id=tokenizer.lang_code_to_id["eng_Latn"])
    decoded_output = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

    # print prompt and translation
    print(input_text)
    print('Translation:')
    print(decoded_output)
    print(dash_line)
```

* This model works well in translation
* It took 5 mins to translate 9 transcripts

### 2.1.2 Task 2: Extractive Question Answering

#### 2.1.2.1 Roberta-Base-Squad2 Model
model info: https://huggingface.co/deepset/roberta-base-squad2

```python
from transformers import pipeline
```

```python
# I used the distilled version: https://huggingface.co/deepset/tinyroberta-squad2
model_name = 'deepset/tinyroberta-squad2'

nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
```

```python
clean_text_id = [3414, 3945, 2526, 178, 3758]
```

```python
def extractive_qa(transcript_id):
  transcript = transcripts[transcripts['id']==id]['transcript'].values[0]
  print(transcript)
  print('')
  for i in range(6):
    question = test[test['Transcript']==id]['Question'].values[i]
    QA_input = {
        'question': question,
        'context': transcript
        }
    res = nlp(QA_input)
    print(f'Question:\n {question}')
    print(f'Answer:\n {res["answer"]}')
  print(dash_line)
```

```python
# test the model with clean text
for id in clean_text_id:
    extractive_qa(id)
```

* It seems that the extractive QA model works well with dictation transcripts, but performs poorly with conversation transcripts.

```python
clean_dictation_id = [3277, 3155, 1613, 4105, 3763, 4271, 3205, 3745, 3775, 4071]
```

```python
# test the model with clean dictation text
for id in clean_dictation_id:
  extractive_qa(id)
```

* The model performs well overall, but sometimes confuses condition and symptoms

```python
messy_text_id = [608, 1848, 1221, 2287, 1766, 3281, 4494]
```

```python
# test the model with messy text
for id in messy_text_id:
    extractive_qa(id)
```

* The model seems robust to messy text - we don't need to clean the text

```python
# test the model with clean text and instructions
for id in clean_text_id:
    transcript = transcripts[transcripts['id']==id]['transcript'].values[0]
    print(transcript)
    print('')
    for i in range(6):
        question = test[test['Transcript']==id]['Question'].values[i]

        prompt = f"""
        In the context, if there are 'D:' and 'P:', then D represents the doctor and P represents the patient, and the context is a conversation between them.
        Otherwise the context is the doctor's dictation.
        If you can't find the information for the question, please answer '?'.
        {question}
        """

        QA_input = {
            'question': prompt,
            'context': transcript
            }
        res = nlp(QA_input)
        print(f'Question:\n {question}')
        print(f'Answer:\n {res["answer"]}')
    print(dash_line)
```

* The Extractive QA model seems not able to comprehend instructions in the prompt



* Conclusions:


1.   Extractive QA models only work well with dictation transcripts
2.   Extractive QA models are somewhat robust to messy text
3.   Extractive QA models are not able to comprehend instructions in the prompt


We can try different extractive QA models and see which one works best on dictation transcripts.
